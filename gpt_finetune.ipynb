#Load model and tokenizer

```python
examples, metadata = tfds.load('ted_hrlr_translate/pt_to_en',
                               with_info=True,
                               as_supervised=True)
```

```python
metadata
```

```python
df = tfds.as_dataframe(examples['train'], metadata)
```

```python
df_valid =  tfds.as_dataframe(examples['validation'], metadata)
```

```python
from datasets import Dataset
import datasets
train_data = Dataset.from_dict(df)
valid_data = Dataset.from_dict(df_valid)
raw_data = datasets.DatasetDict({"train": train_data, "test":valid_data})
```

```python
raw_data
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

```python
pt_sentence = raw_data["train"][1]["pt"].decode()
en_sentence = raw_data["train"][1]["en"].decode()

inputs = tokenizer(pt_sentence, text_target = en_sentence)
inputs
```

```python
max_length = 128


def preprocess_function(examples):
    ens =[s.decode() for s in examples['en']]
    pts = [s.decode() for s in examples['pt']]
    input = [a +' ,in english is ' +b  for a,b in zip(pts,ens)]
    model_inputs = tokenizer(
        input, max_length=max_length, truncation=True
    )
    return model_inputs
```

```python
examples = raw_data['train'][:6]
examples
```

```python
ens =[s.decode() for s in examples['en']]
pts = [s.decode() for s in examples['pt']]
input = [a +' = ' + b for a,b in zip(pts,ens)]

print(input)
model_inputs = tokenizer(
        input, max_length=max_length, truncation=True
    )

print()
print(model_inputs)
```

```python
example = preprocess_function(raw_data['train'][:6])
example
```

```python
tokenizer.decode(example['input_ids'][1])
```

```python
tokenized_datasets = raw_data.map(
    preprocess_function,
    batched=True,
    remove_columns=raw_data["train"].column_names,
)
```

```python
tokenized_datasets
```

```python
from transformers import AutoTokenizer, TFGPT2LMHeadModel, AutoConfig

config = AutoConfig.from_pretrained(
    "gpt2",
    vocab_size=len(tokenizer),
    n_ctx=max_length,
    bos_token_id=tokenizer.bos_token_id,
    eos_token_id=tokenizer.eos_token_id,
)
```

```python
model = TFGPT2LMHeadModel(config)
model(model.dummy_inputs)
model.summary()
```

#Data collator

```python
from transformers import DataCollatorForLanguageModeling

tokenizer.pad_token = tokenizer.eos_token
data_collator = DataCollatorForLanguageModeling(tokenizer, mlm=False)
```

```python
print(tokenized_datasets['train'][0])
```

```python
out = data_collator([tokenized_datasets["train"][i] for i in range(5)])
for key in out:
    print(f"{key} shape: {out[key].shape}")
```

```python
tf_train_dataset = model.prepare_tf_dataset(tokenized_datasets['train'], collate_fn = data_collator, shuffle = True, batch_size = 32)
tf_valid_dataset = model.prepare_tf_dataset(tokenized_datasets['test'], collate_fn = data_collator, shuffle = False, batch_size = 32)
```

#Train model

```python
from transformers import create_optimizer

num_train_steps = len(tf_train_dataset)
opt, scheduler = create_optimizer(init_lr=5e-5,
    num_warmup_steps=0,
    num_train_steps=num_train_steps,
    weight_decay_rate=0.01,)

model.compile(optimizer = opt)
```

```python
num_epochs = 5
model.fit(
    tf_train_dataset,
    validation_data=tf_valid_dataset,
    epochs=num_epochs,
)
```

```python
text = "Ce n'est pas une bonne chose à faire., in english is "
target_text = "This is not a good thing to do."
tokens = tokenizer(text, return_tensors ='tf')
output = model.generate(**tokens, max_length=40)
output
```

```python
import evaluate
metric = evaluate.load("sacrebleu")

import numpy as np
import tensorflow as tf
from tqdm import tqdm

generation_data_collator = DataCollatorForLanguageModeling(tokenizer, mlm=False, return_tensors='tf')

tf_generate_dataset = model.prepare_tf_dataset(
    tokenized_datasets["validation"],
    collate_fn=generation_data_collator,
    shuffle=False,
    batch_size=8,
)



def generate_with_xla(batch):
    return model.generate(
        input_ids=batch["input_ids"],
        attention_mask=batch["attention_mask"],
        max_new_tokens=128,
    )


def compute_metrics():
    all_preds = []
    all_labels = []

    for batch, labels in tqdm(tf_generate_dataset):
        predictions = generate_with_xla(batch)
        decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
        labels = labels.numpy()
        labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
        decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)
        decoded_preds = [pred.strip() for pred in decoded_preds]
        decoded_labels = [[label.strip()] for label in decoded_labels]
        all_preds.extend(decoded_preds)
        all_labels.extend(decoded_labels)

    result = metric.compute(predictions=all_preds, references=all_labels)
    return {"bleu": result["score"]}
```
